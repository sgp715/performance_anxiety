### Reading large local files
The first data we are going to look at was generated by reading in large text files. The functions were called using an increased number of files (each file is 7.3 MB) that they had to process and the time it took them to process each was then recorded.

```python
simple = pd.read_csv("test_fun_dumby.csv")
simple.head()
```

```python
asynchronous = pd.read_csv("test_fun_process_async.csv")
asynchronous.head()
```

```python
threaded = pd.read_csv("test_fun_process_thread.csv")
threaded.head()
```

```python
fig = plt.figure()
ax = plt.gca()
ax.set_xlabel("Files")
ax.set_ylabel("Time")
with pd.plotting.plot_params.use('x_compat', True):
    simple["time"].plot(color='r')
    asynchronous["time"].plot(color='g')
    threaded["time"].plot(color='b')
plt.show()
```

### Insights
So, basically threading sucks(the threading started crashing my apps around 36 files)...Async does ok but seems a little less consistent than sync. Ok, now what if you switch up the tasks and send networking requests. Hopefully this gives some more interesting results.
*******************************************************************************************************************

### Making remote networking requests
The next data we are going to look at was generated by making http requests to a website. The functions were called using with an increased number of requests each time and the time it took to make all the requests is recorded.

```python
simple_requests = pd.read_csv("test_fun_dumby_requests.csv")
simple.head()
```

```python
asynchronous_requests = pd.read_csv("test_fun_async_requests.csv")
asynchronous.head()
```

```python
threaded_requests = pd.read_csv("test_fun_thread_requests.csv")
threaded.head()
```

```python
fig = plt.figure()
ax = plt.gca()
ax.set_xlabel("Requests")
ax.set_ylabel("Time")
with pd.plotting.plot_params.use('x_compat', True):
    simple_requests["time"].plot(color='r')
    asynchronous_requests["time"].plot(color='g')
    threaded_requests["time"].plot(color='b')
plt.show()
```

### Insights
Threading redeems itself here by seeming to be the most consistent and fastest method. Although, with networking there are so many factors it is hard to call this a controlled experiment. Still, the major take aways are that synchronous is the slowest asynce comes in second and threading takes the cake. However, threading also consumes much more memory than the async single-threaded method. So, async is a lightweight and fast option.
